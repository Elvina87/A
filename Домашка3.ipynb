Так же для поиска пустых (NaN) ячеек в датафрейме можно использовать методы библиотеки Pandas. Используем метод -  .isnull(). Метод выполняет одно и то же действие и возвращают булевский датафрейм той же формы, что и исходный, но с True в местах, где значения пустые, и False в местах, где значения не пустые.

Попробуем отфильтровать строчки с пустыми ячейками по столбцам
work_year: год трудоустройства,
зарплата_in_usd: Эквивалентная сумма зарплаты в долларах США (долларах США),
Remote_ratio: процент удаленной работы, разрешенный для данной должности.

```python
DataFrame.isnull().loc[DataFrame['remote_ratio']==True]
```

```python
DataFrame.isnull().loc[DataFrame['salary_in_usd']==True]
```

```python
DataFrame.isnull().loc[DataFrame['work_year']==True]
```

Поиск пустых ячеек подтвердил нам, что их в нашем массиве нет.
